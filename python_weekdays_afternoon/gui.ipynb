```python
from tkinter import *
from tkinter import ttk
from googletrans import LANGUAGES
window=Tk()
window.geometry('1080x480')

window.config(bg='#EFC3CA')
Label(window,text='Language Translator',font='areal 20 bold').place(x=390,y=50)
Label(window,text='Source Language',font='areal 15').place(x=100,y=100)
Label(window,text='Destination Language',font='areal 15').place(x=650,y=100)
#language list contain all languages
language_list=list(LANGUAGES.values())
s=ttk.Combobox(window,values=language_list)   # s: source language variable
s.place(x=270,y=100)
d=ttk.Combobox(window,values=language_list)       # d: destination language variable
d.place(x=850,y=100)
window.title('Language Translator')
Input_text=Text(window,height=15,font='arial 10',width=50)
Input_text.place(x=100,y=150)
Output_text=Text(window,height=15,width=50,font='arial 10')
Output_text.place(x=650,y=150)
trans_but=Button(window,text='Translator',font='arial 10')
trans_but.place(x=500,y=400)
window.mainloop()            # to run windows
```
